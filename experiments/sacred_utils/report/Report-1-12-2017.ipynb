## Current algorithm variation to GSD comparison


In GSD-MDS
- The iterations stop when error starts increasing
- Radius (step) is updated based on the continuous update rule 
\begin{equation}
r_{i+1} = max(s, \frac{10s}{\sqrt{i + 1}})
\end{equation}
- In the old code $s=0.01$ (and generally $s$ takes small values) 
- In each turn the points are sorted based on the error and a percentage of those points are pertubated

In MDS-v2
- We allow the iterations even if error increases temporarily
- Radius (step) is updated based on the rule
\begin{equation*}
r_{i+1} = \begin{cases}
\frac{r_{i}}{2} &\text{if $e{i} - e_{i+1} < c * e_{i}$}\\
r_{i} &\text{otherwise}
\end{cases}\\
\text{where $c$ is an error tolerance coeffient (typically $0.01$) and $r_{0} = 1$}
\end{equation*}
- The iterations stop when radius becomes smaller than a predefined value (e.g. $10^{-3}$)
- In each turn a percentage of randomly sampled points are pertubated
